#### ğŸ“ **Code Examples**: Boxing and Unboxing in Action

```
// === BOXING EXAMPLES ===
// Boxing occurs when a value type is converted to a reference type
// Unboxing occurs when reference type is converted to a value type

object o = 10; 
// â†‘ BOXING: int (value type) â†’ object (reference type)
// The value 10 is allocated on the heap inside an object wrapper

int i = (int)o; 
// â†‘ UNBOXING: object (reference type) â†’ int (value type)
// The boxed value is extracted from the heap back to the stack

int? j = 10; 
// â†‘ BOXING: int (value type) â†’ Nullable<int> (reference-like behavior)
// Nullable<T> can cause boxing when converted to object or interfaces

int k = 10; 
// â†‘ NO BOXING: int (value type) stays as int (value type)
// Direct assignment between same types - no conversion needed

// === MORE DETAILED EXAMPLES ===

// Boxing when calling object methods on value types
int number = 42;
string result = number.ToString(); 
// â†‘ BOXING occurs here! int.ToString() boxes the int to call the method

// Boxing when using interfaces
IComparable comparable = 42; 
// â†‘ BOXING: int â†’ IComparable interface (reference type)

// Boxing in collections that store objects
ArrayList list = new ArrayList();
list.Add(123); 
// â†‘ BOXING: int â†’ object (ArrayList stores objects)

// NO boxing with generic collections
List<int> genericList = new List<int>();
genericList.Add(123); 
// â†‘ NO BOXING: List<T> is type-safe, no conversion needed

// Boxing with string concatenation
string text = "Value: " + 42; 
// â†‘ BOXING: int â†’ object for string.Concat method

// NO boxing with string interpolation (C# 6.0+)
string interpolated = $"Value: {42}"; 
// â†‘ NO BOXING: compiler optimizes this
```

ğŸ“ **Note**:

#### ğŸ“ **Code Examples**: LINQ Method vs Query Syntax

```
// === LINQ TO OBJECTS EXAMPLES ===
// Demonstrating both query syntax and method syntax

var list = new List<int>() { 1, 5, 8, 3, 7, 2, 9 };

// Method syntax - fluent approach
var oddNumbersMethod = list.Where(i => i % 2 == 1).OrderBy(i => i);
Console.WriteLine("Odd numbers (method): " + string.Join(", ", oddNumbersMethod));

// Query syntax - SQL-like approach  
var oddNumbersQuery = from i in list 
                      where i % 2 == 1 
                      orderby i 
                      select i;
Console.WriteLine("Odd numbers (query): " + string.Join(", ", oddNumbersQuery));

// Complex example with chaining
var processedData = list
    .Where(x => x > 2)           // Filter
    .Select(x => x * 2)          // Transform
    .OrderByDescending(x => x)   // Sort
    .Take(3);                    // Limit
Console.WriteLine("Processed: " + string.Join(", ", processedData));
```
