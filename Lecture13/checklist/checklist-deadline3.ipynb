### üö´ **Critical Issues (Major Point Deductions)**

- **Hardcoded credentials**: Database connection strings, API keys, passwords in source code (-0.5 points)

- **No input validation**: SQL injection, XSS vulnerabilities in production code (-0.4 points)
- **Insecure authentication**: Plain text passwords, weak session management (-0.3 points)

#### **Code Quality & Stability Issues**
- **Unhandled exceptions**: Application crashes during presentation due to unhandled errors (-0.3 points)
- **Broken main branch**: Non-functional code in main/master branch during presentation (-0.5 points)
- **Non-working features**: Demonstrating features that don't actually work (-0.4 points)

- **Direct commits to main**: Bypassing pull request process for main branch (-0.2 points)
- **Meaningless commit messages**: "fix", "update", "asdf" type commit messages (-0.1 points) 
  *Note: One or two commits with non-descriptive messages will not be punished, but this should not be a common occurrence throughout the project*
- **No code reviews**: Pull requests without meaningful team member reviews (-0.2 points) 
  *Note: One or two pull requests without meaningful reviews will not be heavily penalized, but lack of code review should not be a common pattern throughout the project*
- **Inactive team members**: Team members who can't explain their contributions (-0.3 points per member)

### ‚ö†Ô∏è **Moderate Issues (Minor Point Deductions)**
